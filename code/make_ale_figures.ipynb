```python
%matplotlib inline
import matplotlib.pyplot as plt
from nilearn import plotting
from glob import glob
import os
```

```python
in_dir = os.path.abspath("../derivatives/ale")
out_dir = os.path.abspath("../derivatives/ale-figures")

os.makedirs(out_dir, exist_ok=True)

files = sorted(glob(os.path.join(in_dir, "*z_corr-FWE_thresh-05*")))
print(files)
```

```python
for f in files:
    name = os.path.basename(f)
    name = name.split("_")[0]
    fig, ax = plt.subplots(figsize=(16, 4))
    plotting.plot_stat_map(
        f,
        display_mode="x",
        cut_coords=[-8, -4, 0, 4],
        cmap="Purples",
        figure=fig,
        axes=ax,
    )
    # ax.set_title(name)

    fig.savefig(
        os.path.join(out_dir, name + ".svg"),
        transparent=True,
        bbox_inches="tight",
        pad_inches=0,
    )
```
